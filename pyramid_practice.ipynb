```python
    im_path = 'tsukuba_l.png'
    im_src = cv2.imread(im_path, cv2.IMREAD_GRAYSCALE) 
    print(im_src.shape)
    plt.imshow(im_src,cmap = "gray")
```

```python
#pyrDown ~ 피라미드 
im2=im_src[::2,::2]  #[::2,::2,::] 는 3차원
print(im2.shape)
plt.imshow(im2,cmap="gray")
```

```python
#샘플링 여러번!

im2=im_src[::2,::2] 
im3=im2[::2,::2] 
im4=im3[::2,::2] 
im5=im4[::2,::2] 


print(im5.shape)
plt.imshow(im5,cmap="gray")
```

```python
plt.imshow(im4, cmap = "gray")
```

```python
#다운사이징

im2=cv2.pyrDown(im_src)
im3=cv2.pyrDown(im2)
im4=cv2.pyrDown(im3)
im5=cv2.pyrDown(im4)

print(im5.shape)
plt.imshow(im5, cmap = "gray")
```

```python
plt.imshow(im4, cmap = "gray")

#덜 중요한 정보는 사라진 상태 
```

```python
#업샘플링

im4u=cv2.pyrUp(im5)
im3u=cv2.pyrUp(im4)
im2u=cv2.pyrUp(im3)
im1u=cv2.pyrUp(im2)


# 다운샘플링 하고 다시 없샘플링 한 이미지는 원본과 동일하지 않는다!!!
#원본이미지 : LF + HF
#다운샘플링 : LF만 전달된다. (HF는 사라진다)
```

```python
print(im3u.shape)
plt.imshow(im3u, cmap = "gray")
```

```python
#라플라시안 이미지 만들기 
#해당이미지를 다운샘플링 하여 다시 업샘플링을 빼주면 
#현재 이미지에서 HF(변화가 빨리 나타나는 부분 = 경계)만 남아서 출력됨. 


im4l= cv2.subtract(im4,im4u)
im3l= cv2.subtract(im3,im3u)
im2l= cv2.subtract(im2,im2u)
im1l= cv2.subtract(im_src,im1u)
```
